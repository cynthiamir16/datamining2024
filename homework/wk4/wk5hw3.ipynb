```python
5. Brightness Adjustment

● Write a script to increase the brightness of the image by a certain value and display the brightened image.
```

```python
image_float32 = p.astype(np.float32)
brightness_factor = 50
brightened = image_float32 * (1 + brightness_factor / 100.0)
brightened = np.clip(brightened, 0, 255)
bright = brightened.astype(np.uint8)
plt.imshow(cv2.cvtColor(bright, cv2.COLOR_BGR2RGB))
```

```python
6. Blurring and Smoothing:

● Apply Gaussian Blur to 'Pandas.png' and display both the original and blurred images.

● Try different kernel sizes for blurring on 'Pandas.png' and discuss the effects.
```

```python
blurred_image = cv2.GaussianBlur(p, (13, 13), 0)
plt.figure(figsize=(11, 11))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(p, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
7. Image Indexing and Cropping:

● Crop a portion of the 'Pandas.png' image using indexing and display the cropped image.

● Change the color of a specific region of 'Pandas.png' using indexing and display the result.
```

```python
top_left = (600, 600)
bottom_right = (1000, 1000)

cropped_image = p[top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]]

plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
```

```python
mask = np.zeros(p.shape[:2], dtype=np.uint8)
cv2.rectangle(mask, top_left, bottom_right, (255, 255, 255), -1)
p[mask > 0] = (0, 255, 0)  # Set the color to green
plt.imshow(cv2.cvtColor(p, cv2.COLOR_BGR2RGB))
```

```python
8. Image Rotation:

● Rotate the 'Pandas.png' image by 45 degrees without cropping the corners and display the result.

● Discuss the challenges and
```

```python
height, width = p.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), 45, 1)
rotated_image = cv2.warpAffine(p, rotation_matrix, (width, height))
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
```

```python
11. Gradients in Image:

● Calculate and display the gradient direction and magnitude images
```

```python
x = cv2.Sobel(p, cv2.CV_64F, 1, 0, ksize=3)
y = cv2.Sobel(p, cv2.CV_64F, 0, 1, ksize=3)


gradient_magnitude = np.sqrt(np.square(x) + np.square(y))
gradient_direction = np.arctan2(y, x)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gradient_magnitude, cmap='gray')
```
